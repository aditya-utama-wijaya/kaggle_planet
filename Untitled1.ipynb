```python
from warnings import filterwarnings
filterwarnings('ignore')

import pandas as pd
pd.set_option('display.max_columns', None)

import vgg16
from data_helper import AmazonPreprocessor

labels_df = pd.read_csv('train_v2.csv')
print(labels_df.shape)
labels_df.tail()
```

```python
from itertools import chain

labels_list = list(chain.from_iterable([t.split(' ') for t in labels_df['tags'].values]))
labels_set = set(labels_list)
print(len(labels_set))
labels_set
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

labels_s = pd.Series(labels_list).value_counts()
fig, ax = plt.subplots(figsize = (16, 8))
sns.barplot(x = labels_s, y = labels_s.index, orient = 'h')
```

```python
import matplotlib.image as mpimg

images_title = [labels_df[labels_df['tags'].str.contains(label)].iloc[i]['image_name'] + '.jpg' for i, label in enumerate(labels_set)]

plt.rc('axes', grid = False)
_, axs = plt.subplots(5, 4, sharex = 'col', sharey = 'row',
                      figsize = (15, 20))
axs = axs.ravel()

for i, (image_name, label) in enumerate(zip(images_title, labels_set)):
    img = mpimg.imread(f'train-jpg/{image_name}')
    axs[i].imshow(img)
    axs[i].set_title(f'{image_name} - {label}')
```

```python
import data_helper

img_resize = (128, 128)
validation_split_size = 0.2
train_jpeg_dir, train_csv_file = data_helper.get_jpeg_data_files_paths()

preprocessor = AmazonPreprocessor(train_jpeg_dir,
                                  train_csv_file, 
                                  img_resize,
                                  validation_split_size)
preprocessor.init()
```

```python
print("X_train/y_train length: {}/{}".format(len(preprocessor.X_train), len(preprocessor.y_train)))
print("X_val/y_val length: {}/{}".format(len(preprocessor.X_val), len(preprocessor.y_val)))
preprocessor.y_map
```

```python
model = vgg16.create_model(img_dim = (128, 128, 3))
model.summary()
```

```python
from keras.optimizers import Adam

X_train, y_train = preprocessor.X_train, preprocessor.y_train
X_val, y_val = preprocessor.X_val, preprocessor.y_val

batch_size = 128
train_generator = preprocessor.get_train_generator(batch_size)
steps = len(X_train) / batch_size

model.compile(optimizer = Adam(lr = 1e-4),
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
history = model.fit_generator(train_generator, 
                              steps,
                              epochs = 25,
                              verbose = 1,
                              validation_data = (X_val, y_val))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'validation'], loc = 'upper left')
```

```python
model.load_weights('weights/weights.best.hdf5')
print('Weights loaded')
```

```python
fbeta_score = vgg16.fbeta(model, X_val, y_val)
fbeta_score
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau, History

history = History()
callbacks = [history,
             EarlyStopping(monitor = 'val_loss',
                           patience = 3, 
                           verbose = 1,
                           min_delta = 1e-4),
             ReduceLROnPlateau(monitor = 'val_loss', 
                               factor = 0.1,
                               patience = 1,
                               cooldown = 0,
                               min_lr = 1e-7, 
                               verbose = 1),
             ModelCheckpoint(filepath = 'weights/weights.best.hdf5',
                             verbose = 1,
                             save_best_only = True,
                             save_weights_only = True,
                             mode = 'auto')]

model.compile(optimizer = Adam(lr = 1e-4),
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])
history = model.fit_generator(train_generator, 
                              steps,
                              epochs = 25,
                              verbose = 1,
                              validation_data = (X_val, y_val),
                              callbacks = callbacks)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['train', 'validation'], loc = 'upper left')
```

```python
model.load_weights('weights/weights.best.hdf5')
print('Weights loaded')
```
